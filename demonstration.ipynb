```python
from Protein import Protein
import pandas as pd
import numpy as np
```

```python
#Loading the dataframe to get protein sequences
df = pd.read_csv('./data/prokaryotes_348columns.csv')
sequences = df['Sequence']
sequences.index = df['AlphaFoldDB'].str[:-1]
# Example Protein 1 Exodeoxyribonuclease III (P09030)
seq1 = sequences['P09030']
# Example Protein 2 HTH-type transcriptional regulator GltR (P94501)
seq2 = sequences['P94501']
```

```python
# Creating Protein objects with sequences as mandatory arguments and PDBs as optional arguments required for mutations
Nuclease = Protein(seq = seq1, PDB = 'AF-P09030-F1.pdb')
GltR = Protein(seq = seq2, PDB = 'AF-P94501-F1.pdb')
# all features were calculated in the __init__ method and stored as attributes
```

```python
# predicted melt points
print(Nuclease.SPARC)
print(GltR.SPARC)
# other features
print(Nuclease.Helix2count)         # Number of helices
print(GltR.PolarAA)                 # Relative amount of polar amino acids
print(Nuclease.Rhelix)              # Relative helix arginine content
print(GltR.ER)                      # Relative glutamic acid + arginine content
print(GltR.EAmotif)                 # Relative amount of EA dipeptides
all_features = Nuclease.features    # Series with all features
```

```python
# using ThERMOS to mutate a protein
Nuclease.ThERMOS()
```

```python
print(f'WT prediction:{Nuclease.SPARC}\nMutation prediction:{Nuclease.mutatedmp}')
print(Nuclease.sequence)            # Original sequence
print("".join(Nuclease.mutatedseq)) # Mutated sequence
print(Nuclease.mutationlist)        # List of mutations
print(len(Nuclease.mutationlist))   # Number of mutations
```

```python
# using ThERMless to reduce the number of mutations while keeping as much of a melt point increase as possible
Nuclease.ThERMless()
```

```python
print(Nuclease.removed_mutations) # reverted mutations
print(Nuclease.mut_mp_reduced)    # melt point of reduced mutated protein
```
